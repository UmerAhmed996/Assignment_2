```python
import pandas as pd

df = pd.read_csv("metadata.csv")

import numpy as np

missingdata = df.isnull()

missingdata.head()
```

```python
missingdata.sum()
```

```python
# Removing missing data

avg = df["subtlety"].astype("float").mean(axis=0)
df["subtlety"].replace(np.nan,avg,inplace = True)
df["subtlety"].isnull().sum()

avg = df["size"].astype("float").mean(axis=0)
df["size"].replace(np.nan,avg,inplace = True)
df["size"].isnull().sum()

avg = df["x"].astype("float").mean(axis=0)
df["x"].replace(np.nan,avg,inplace = True)
df["x"].isnull().sum()

avg = df["y"].astype("float").mean(axis=0)
df["y"].replace(np.nan,avg,inplace = True)
df["y"].isnull().sum()

m = df["position"].value_counts().idxmax()
df["position"].replace(np.nan, m, inplace=True)
df["position"].isnull().sum()

m = df["diagnosis"].value_counts().idxmax()
df["diagnosis"].replace(np.nan, m, inplace=True)
df["diagnosis"].isnull().sum()
```

```python
missingdata = df.isnull()
missingdata.sum()
# Missing Data Removed Successfuly

m = df["gender"].value_counts()
m = m.astype("float")

print('Total Number of Patients with the state are following : \n',m)
```
