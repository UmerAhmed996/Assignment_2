```python
import pandas as pd
import numpy as np

FileData = pd.read_csv("metadata.csv")
miss = FileData.isnull()
```

```python
miss.sum()
```

```python
A = FileData["size"].astype("float").mean(axis=0)
FileData["size"].replace(np.nan,A,inplace = True)

A = FileData["subtlety"].astype("float").mean(axis=0)
FileData["subtlety"].replace(np.nan,A,inplace = True)

A = FileData["y"].astype("float").mean(axis=0)
FileData["y"].replace(np.nan,A,inplace = True)

A = FileData["x"].astype("float").mean(axis=0)
FileData["x"].replace(np.nan,A,inplace = True)

M = FileData["position"].value_counts().idxmax()
FileData["position"].replace(np.nan, M, inplace=True)

M = FileData["diagnosis"].value_counts().idxmax()
FileData["diagnosis"].replace(np.nan, M, inplace=True)

miss = FileData.isnull()
miss.sum()
```
